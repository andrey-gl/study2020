#Математическое ожидание, дисперсия, ковариация

Математическое ожидание - это, говоря не строго, среднее значение, которое принимает случайная величина. Так, среднее значение, выпадающее на игральной кости, равно 3.5, и математическое ожидание случайной величины, равной значению, выпавшему на кости, также равно 3.5. Если случайная величина имеет дискретное распределение, то определение математического ожидание вводится естественным образом как взвешенная сумма значений случайной величины, и мы его приводим ниже.

**Определение.** Математическим ожиданием случайной величины $\xi$, имеющей дискретное распределение,} называется величина ${\sf E}\xi$, равная $\sum_{x\in X}x{\sf P}(\xi=x)$, где $X$ - множество значений $\xi$.

бесконечна

Если $\varphi:\mathbb{R}\rightarrow\mathbb{R}$ --- произвольная
функция, то ${\sf E}\varphi(\xi)$ может быть вычислено по формуле
$$
{\sf E}\varphi(\xi)=\sum_{x\in X}\varphi(x){\sf P}(\xi=x),
 $$
где $X$ - множество значений случайной величины
$\xi$.


Посчитаем математические ожидания случайных величин, имеющих
различные дискретные распределения.


1. Пусть $\xi\sim Bern(p)$, то есть ${\sf P}(\xi=1)=1-{\sf
 P}(\xi=0)=p$. Тогда ${\sf E}\xi=p$.

2. Пусть $\xi\sim U\{1,\ldots,n\}$, то есть ${\sf P}(\xi=i)=\frac{1}{n},$ $i\in\{1,\ldots,n\}$.
 Тогда ${\sf E}\xi=\frac{1}{n}\sum_{i=1}^n
 i=\frac{n+1}{2}$.

3. Пусть $\xi\sim Pois(\lambda)$, то есть ${\sf
 P}(\xi=k)=\frac{e^{-\lambda}\lambda^k}{k!}$. Тогда ${\sf E}\xi=\sum_{k=0}^{\infty}k{\sf P}(\xi=k)=
 \lambda
 e^{-\lambda}\sum_{k=1}^{\infty}\frac{\lambda^{k-1}}{(k-1)!}=\lambda$.

Перед тем, как перейти к общему случаю, дадим определение
математического ожидания в абсолютно непрерывном случае.


**Определение.** *Математическим ожиданием случайной
величины $\xi$, имеющей абсолютно непрерывное распределение*,
называется величина ${\sf E}\xi$, равная
$\int_{-\infty}^{\infty}xp_{\xi}(x)dx$.

Заметим, что абсолютно непрерывный случай является естественным
обобщением дискретного, так как ${\sf P}(\xi\in(x,x+\Delta
x))\approx p_{\xi}(x)\Delta x$.

Посчитаем математические ожидания случайных величин, имеющих
различные абсолютно непрерывные распределения.

1. Пусть $\xi\sim Exp(\lambda)$, то есть $p_{\xi}(x)=\lambda e^{-\lambda x}I(x>0)$.
 Тогда, воспользовавшись формулой интегрирования по частям,
 получаем
 $$
  {\sf E}\xi=\int_0^{\infty}\lambda x e^{-\lambda x}dx=
  -\int_0^{\infty}x d\left(e^{-\lambda x}\right)=
  \int_0^{\infty}e^{-\lambda x}dx=\frac{1}{\lambda}.
 $$

2. Пусть $\xi\sim \mathcal{N}(a,\sigma^2)$, то есть $p_{\xi}(x)=\frac{1}{\sqrt{2\pi\sigma^2}}e^{-\frac{(x-a)^2}{2\sigma^2}}$.
 Тогда
 $$
  {\sf E}\xi=\int_{-\infty}^{\infty}\frac{x}{\sqrt{2\pi\sigma^2}}
  e^{-\frac{(x-a)^2}{2\sigma^2}}dx=
  \int_{-\infty}^{\infty}\frac{x-a+a}{\sqrt{2\pi\sigma^2}}e^{-\frac{(x-a)^2}{2\sigma^2}}dx=
  \sqrt{\frac{\sigma^2}{2\pi}}\int_{-\infty}^{\infty}
  e^{-\frac{(x-a)^2}{2\sigma^2}}d\left(\frac{(x-a)^2}{2\sigma^2}\right)+a=a,
 $$
 так как $\int_{-\infty}^{\infty}p_{\xi}(x)dx=1$.

В общем случае математическим ожиданием случайной величины $\xi$
называется интеграл Лебега $\int_{\Omega}\xi d{\sf P}$. Дадим
определение интеграла Лебега по вероятностной мере ${\sf P}$.

Важным свойством математического ожидания является свойство
линейности.

**Теорема.** *Если ${\sf E}\xi$ существует, то ${\sf
E}(c\xi)=c{\sf E}\xi$. Если существуют и конечны ${\sf E}\xi,{\sf
E}\eta,$ то существует и ${\sf E}(\xi+\eta),$ причем ${\sf
E}(\xi+\eta)={\sf E}\xi+{\sf E}\eta$.*

Как мы знаем, если $\xi$ - случайная величина, а $\varphi$ -
борелевская функция, то и $\varphi(\xi)$ - случайная величина.
Предположим, что мы знаем распределение случайной величины $\xi$.
Как вычислить ${\sf E}\varphi(\xi)$? Разумеется, можно найти
распределение $\varphi(\xi)$, после чего воспользоваться
определением математического ожидания. Однако, проще
воспользоваться теоремой о замене переменных в интеграле Лебега.\\

**Теорема.** Пусть $\xi$ - случайная величина, а
$\varphi$ - борелевская функция. Если существует ${\sf
E}\varphi(\xi)$, то
$$
 \int_{\Omega}\varphi(\xi(\omega)){\sf
 P}(d\omega)=\int_{\mathbb{R}}\varphi(x){\sf P}_{\xi}(dx),
$$
*где ${\sf P}_{\xi}$ - распределение вероятностей случайной
величины $\xi$.

**Задача.** Пусть $\xi\sim Pois(\lambda)$. Вычислите ${\sf
E}e^{\xi}$.

**Решение.** Имеем
$$
 {\sf
 E}e^{\xi}=\sum_{k=0}^{\infty}e^{k}\frac{\lambda^ke^{-\lambda}}{k!}
 =e^{-\lambda}\sum_{k=0}^{\infty}\frac{(e\lambda)^k}{k!}
 =e^{\lambda(e-1)}.
 \,\,\,\Box
$$


Разумеется, по одному известному среднему значению случайной
величины сложно судить о ее распределении. Дополнительную
информацию дает среднее квадратическое отклонение от
математического ожидания, которое называют дисперсией.

**Определение.** *Дисперсией* случайной величины $\xi$
называется величина ${\sf E}(\xi-{\sf E}\xi)^2$. *Ковариацей*
случайных величин $\xi,\eta$ называется
$cov(\xi,\eta)={\sf E}((\xi-{\sf E}\xi)(\eta-{\sf E}\eta))$.

Иными словами, дисперсия это среднее квадратическое отклонение от
среднего. Дробь $\rho(\xi,\eta)=\frac{cov(\xi,\eta)}{\sqrt{{\sf
D}\xi{\sf D}\eta}}$ является, в некотором смысле, мерой
зависимости случайных величин $\xi,\eta$. Так, если случайные
величины $\xi,\eta$ независимы, то $\rho(\xi,\eta)=0$. Если же
$\rho(\xi,\eta)=\pm 1$, то случайные величины линейно зависимы.
Величина $\rho(\xi,\eta)$ называется *коэффициентом корреляции
случайных величин $\xi$ и $\eta$*.

**Свойства.**

1. ${\sf D}\xi={\sf E}\xi^2-({\sf E}\xi)^2.$

2. ${\sf D}\xi\geq 0$. 

3. $cov(\xi,\eta)={\sf E}(\xi\eta)-{\sf E}\xi{\sf E}\eta$.

4. Если $\xi,\eta$ независимы, то $cov(\xi,\eta)=0$.

5. Ковариация билинейна:
$cov(a_1\xi_1+a_2\xi_2,\eta)=a_1cov(\xi_1,\eta)+a_2cov(\xi_2,\eta)$,
$cov(\xi,a_1\eta_1+a_2\eta_2)=a_1cov(\xi,\eta_1)+a_2cov(\xi,\eta_2)$.

6. ${\sf D}(\xi_1+...+\xi_n)={\sf D}\xi_1+...+{\sf
D}\xi_n+\sum_{i\neq j}cov(\xi_i,\xi_j)$.

7. Если $\xi_1,...,\xi_n$ --- попарно независимые случайные
величины, то $D(\xi_1+...+\xi_n)=D(\xi_1)+...+D(\xi_n)$.

# Неравенства на средние

Для доказательства асимпотических свойств последовательностей
случайных величин удобно использовать
неравенства, позволяющие необходимые величины оценивать.

**Неравенство Коши-Буняковского.** *Пусть случайных величины $\xi$ и
$\eta$, заданные на одном и том же вероятностном пространстве
$(\Omega,\mathcal{F},{\sf P})$, таковы, что ${\sf E}\xi^2<\infty$,
${\sf E}\eta^2<\infty$. Тогда ${\sf E}|\xi\eta|<\infty$ и $({\sf
E}|\xi\eta|)^2\leq {\sf E}\xi^2{\sf E}\eta^2.$*


**Неравенство Маркова.** *Пусть $\xi$ --- случайная величина
с конечным математическим ожиданием на вероятностном пространстве
$(\Omega,\mathcal{F},{\sf P})$. Тогда для любого $a>0$ ${\sf
P}(|\xi|\geq a)\leq \frac{{\sf E} |\xi|}{a}$.*

**Неравенство Чебышева.** Случайная величина $\xi$ с
конечным вторым моментом $({\sf E}\xi^2<\infty)$ задана на
вероятностном пространстве $(\Omega,\mathcal{F},{\sf P})$. Тогда
для любого $a>0$ ${\sf P}(|\xi-{\sf E}\xi|\geq a)\leq \frac{{\sf
D}
\xi}{a^2}$.}

Последнее неравенство позволяет получить, так называемый, закон
больших чисел в форме Чебышева. Смысл в том, что при $n$
независимых испытаниях значение суммы близко к $n$ математическим
ожиданием одной случайной величины (пример с монеткой или
игральной костью).

**Закон больших чисел в форме Чебышева.** Пусть
$\xi_1,\xi_2,...$ - независимые одинаково распределенные
случайные величины, заданные на одном и том же вероятностном
пространстве. Пусть, кроме того, ${\sf E}\xi_1^2<\infty$. Тогда
$\forall \varepsilon>0,\delta>0$ ${\sf P}(\frac{|S_n-{\sf
E}S_n|}{n}>\varepsilon)\rightarrow
0.$

$\triangleright$ Достаточно воспользоваться неравенством Чебышева.
Так,
$$
 {\sf P}\left(\frac{|S_n-{\sf
 E}S_n|}{n}>\varepsilon\right)\leq \frac{{\sf
 D}S_n}{n^2\varepsilon^2}=\frac{{\sf
 D}\xi_1}{\varepsilon^2n}\rightarrow 0.\quad \triangleleft
$$

**Замечание.** Если для случайных величин $\xi,\xi_1,
\xi_2,...$ и для любого $\varepsilon>0$ выполнено ${\sf
P}(|\xi_n-\xi|>\varepsilon)\rightarrow 0$, то говорят, что
$\xi_n\rightarrow \xi$ по вероятности (подробности далее).

**Замечание.** Для сходимости по вероятности достаточно существования первого момента ${\sf E}|\xi_1|<\infty$ (ЗБЧ в форме Хинчина).


# Виды сходимостей, центральная предельная теорема

**Определение.** Последовательность случайных величин
$\xi_1,\xi_2,\ldots$ сходится *почти наверное* (или *с
вероятностью 1*) к случайной величине $\xi$ (обозначение
$\xi_n\stackrel{\mbox{п.н.}}{\longrightarrow}\xi$), если ${\sf
P}(\xi_n\rightarrow\xi)=1$.

**Определение.** Последовательность случайных величин
$\xi_1,\xi_2,\ldots$ сходится *по вероятности* к случайной
величине $\xi$ (обозначение $\xi_n\stackrel{{\sf
P}}{\longrightarrow}\xi$), если для любого $\varepsilon>0$
выполнено
${\sf P}(|\xi_n-\xi|\geq\varepsilon)\rightarrow 0$, $n\rightarrow\infty$.

**Определение.** Пусть $p>0$ - произвольное число.
Последовательность случайных величин $\xi_1,\xi_2,\ldots$ сходится
{\it в $L_p$} (обозначение
$\xi_n\stackrel{L_p}{\longrightarrow}\xi$), если ${\sf
E}|\xi_n-\xi|^p\rightarrow 0$,
$n\rightarrow\infty$.

**Определение.** Последовательность случайных величин
$\xi_1,\xi_2,\ldots$ сходится *слабо* (или {*по
распределению*) к случайной величине $\xi$ (обозначение
$\xi_n\stackrel{d}{\longrightarrow}\xi$), если для любой
непрерывной ограниченной функции
$f:\mathbb{R}\rightarrow\mathbb{R}$ выполнено ${\sf E}f(\xi_n)\rightarrow{\sf E}f(\xi)$, $n\rightarrow\infty$.

**Теорема (о связях между различными видами сходимости).**
*Пусть $\xi,\xi_1,\xi_2,\ldots$ - произвольные случайные
величины. Из $\xi_n\stackrel{\mbox{п.н.}}{\longrightarrow}\xi$
следует $\xi_n\stackrel{{\sf P}}{\longrightarrow}\xi$; из
$\xi_n\stackrel{{\sf P}}{\longrightarrow}\xi$ следует
$\xi_n\stackrel{d}{\longrightarrow}\xi$. Кроме того, для любого
$p>0$ из $\xi_n\stackrel{L_p}{\longrightarrow}\xi$ следует
$\xi_n\stackrel{{\sf P}}{\longrightarrow}\xi$.*

**Теорема (центральная предельная теорема для одинаково
распределенных независимых случайных величин).** *Пусть
$\xi_1,\xi_2,\ldots$ - независимые одинаково распределенные
случайные величины с конечным вторым моментом. Тогда
$\frac{S_n-{\sf E}S_n}{\sqrt{{\sf
D}S_n}}\stackrel{d}\longrightarrow\eta$, где
$\eta\sim\mathcal{N}(0,1)$.*
